#### **CARGA DE DATOS**

```python
import pandas as pd
```

```python
dataset_xlsx = pd.ExcelFile('datasets\Telefonia_movil.xlsx')
```

```python
df_sms = pd.read_excel(dataset_xlsx, sheet_name='SMS')
df_llamadas_salientes = pd.read_excel(dataset_xlsx, sheet_name='Llamadas salientes')
df_minutos_salientes = pd.read_excel(dataset_xlsx, sheet_name='Minutos salientes')
df_ingresos = pd.read_excel(dataset_xlsx, sheet_name='Ingresos')
df_penetracion = pd.read_excel(dataset_xlsx, sheet_name='Penetracion')
df_accesos = pd.read_excel(dataset_xlsx, sheet_name='Accesos')
```

---

### **``df_sms``**: número de SMS salientes

```python
df_sms.info()
df_sms.describe()
```

### **``df_llamadas_salientes``**: Cantidad de minutos consumidos en llamadas salientes.

```python
df_llamadas_salientes.info()
df_llamadas_salientes.describe()
```

### **``df_minutos_salientes``**: Cantidad de minutos consumidos en llamadas salientes.

```python
df_minutos_salientes.info()
df_minutos_salientes.describe()
```

### **``df_ingresos``**: Datos de ingresos generados por el servicio de telefonía móvil.

```python
df_ingresos.info()
df_ingresos.describe()
```

### **``df_penetracion``**: Información sobre la penetración del servicio en la población.

```python
df_penetracion.info()
df_penetracion.describe()
```

### **``df_accesos``**: Número de accesos al servicio.

```python
df_accesos.info()
df_accesos.describe()
```

---

#### **EXPORTACIÓN DE DATOS**

##### Se exportan todos los datasets relacionados con **Telefonía Móvil** a archivos CSV para integrarlos a la base de datos.




```python
# Se exportan todos los DataFrames a CSV para carga en SQL

df_sms.to_csv('sms_salientes.csv', index=False)
df_llamadas_salientes.to_csv('llamadas_salientes.csv', index=False)
df_minutos_salientes.to_csv('minutos_salientes.csv', index=False)
df_ingresos.to_csv('ingresos_telefonia_movil.csv', index=False)
df_penetracion.to_csv('penetracion_telefonia_movil.csv', index=False)
df_accesos.to_csv('accesos_telefonia_movil.csv', index=False)
```
